```python
#linhas nan
linhas_nan = tabela_satisfacao_log[tabela_satisfacao_log.isna().all(axis=1)]
linhas_nan
```

```python
#drop linhas nan
tabela_satisfacao_log = tabela_satisfacao_log.dropna(how='all')
tabela_satisfacao_log
```

```python
#colunas nan
colunas_nan = tabela_satisfacao_log.columns[tabela_satisfacao_log.isna().all()]
colunas_nan
```

```python
#drop colunas nan
tabela_satisfacao_log = tabela_satisfacao_log.drop(columns=colunas_nan)
tabela_satisfacao_log
```

```python
tabela_satisfacao_log.isnull().sum()
```

```python
tabela_satisfacao_log.duplicated().sum()
```

```python
#id em string
tabela_satisfacao_log['ticket_id'] = tabela_satisfacao_log['ticket_id'].fillna(0).astype(int)
```

```python
#coluna 'updated_at' para datetime
tabela_satisfacao_log['updated_at'] = pd.to_datetime(tabela_satisfacao_log['updated_at'].str.replace(' UTC', ''))
tabela_satisfacao_log
```

```python
tabela_satisfacao_log.dtypes
```

```python
tabela_satisfacao_log['score'].unique()
```

```python
# renomeando e reordenando
score_traducao = {'good': 'positivo', 'offered': 'enviado', 'bad': 'negativo'}
tabela_satisfacao_log['score'] = tabela_satisfacao_log['score'].replace(score_traducao)
tabela_satisfacao_log = tabela_satisfacao_log.rename(columns={'updated_at': 'data_atualizacao'})
tabela_satisfacao_log = tabela_satisfacao_log.reindex(columns=['ticket_id', 'score', 'data_atualizacao'])
```

```python
tabela_satisfacao_log
```

```python
# últimas datas de atualização para cada ticket_id
ultimos_indices = tabela_satisfacao_log.groupby('ticket_id')['data_atualizacao'].idxmax()

# filtro
tabela_satisfacao_log = tabela_satisfacao_log.loc[ultimos_indices]

tabela_satisfacao_log
```

##Análise


```python
# Juntar as tabelas usando merge
dados_ultima_data = pd.merge(tabela_satisfacao_log, tabela_ticket_metricas, on='ticket_id')
```

```python
dados_ultima_data
```

```python
# coluna intervalo de tempo de resposta até 3 dias ou + de 3 dias
dados_ultima_data['intervalo_tempo'] = pd.cut(dados_ultima_data['tempo_de_resposta'], bins=[-1, 3*24*60, float('inf')], labels=['Até 3 dias', 'Mais de 3 dias'])
dados_ultima_data
```

```python
#apenas os tickets que receberam alguma avaliação
dados_avaliados = dados_ultima_data.dropna(subset=['score'])


positivos = (dados_avaliados['score'] == 'positivo').sum()
negativos = (dados_avaliados['score'] == 'negativo').sum()

proporcao_positivos = positivos / (positivos + negativos)
proporcao_negativos = negativos / (positivos + negativos)

print("Proporção de tickets negativo:", proporcao_negativos)
print("Proporção de tickets positivos:", proporcao_positivos)
```

REFAZER ESTA PARTE-

```python
#satisfação "positivo" e "negativo" em cada intervalo
satisfacao_media = dados_ultima_data.pivot_table(index='intervalo_tempo', columns='score', aggfunc='size', fill_value=0)
satisfacao_media['satisfacao_media_positivo'] = satisfacao_media['positivo'] / (satisfacao_media['positivo'] + satisfacao_media['negativo'])
satisfacao_media['satisfacao_media_negativo'] = satisfacao_media['negativo'] / (satisfacao_media['positivo'] + satisfacao_media['negativo'])
satisfacao_media.reset_index(inplace=True)
satisfacao_media[['intervalo_tempo', 'satisfacao_media_positivo', 'satisfacao_media_negativo']]
```

```python
plt.bar(satisfacao_media['intervalo_tempo'], satisfacao_media['satisfacao_media_positivo'], color='orange', label='Positivo')
plt.bar(satisfacao_media['intervalo_tempo'], satisfacao_media['satisfacao_media_negativo'], color='purple', label='Negativo', bottom=satisfacao_media['satisfacao_media_positivo'])
plt.xlabel('Intervalo de Tempo de Resposta')
plt.ylabel('Satisfação Média')
plt.title('Satisfação Média por Intervalo de Tempo de Resposta')
plt.legend()
plt.show()
```

##Intestigação até 1 dia


```python
limites_intervalos = [-1, 1*24*60, 3*24*60, float('inf')]

rotulos_intervalos = ['Até 1 dia', 'Até 3 dias', 'Mais de 3 dias']

dados_ultima_data['intervalo_tempo'] = pd.cut(dados_ultima_data['tempo_de_resposta'], bins=limites_intervalos, labels=rotulos_intervalos)
dados_ultima_data
```

```python
#média satisfação até 1 dia, até 3 dias e + de 3 dias
satisfacao_media = dados_ultima_data.pivot_table(index='intervalo_tempo', columns='score', aggfunc='size', fill_value=0)
satisfacao_media['satisfacao_media_positivo'] = satisfacao_media['positivo'] / (satisfacao_media['positivo'] + satisfacao_media['negativo'])
satisfacao_media['satisfacao_media_negativo'] = satisfacao_media['negativo'] / (satisfacao_media['positivo'] + satisfacao_media['negativo'])
satisfacao_media.reset_index(inplace=True)
satisfacao_media[['intervalo_tempo', 'satisfacao_media_positivo', 'satisfacao_media_negativo']]
```

```python
plt.bar(satisfacao_media['intervalo_tempo'], satisfacao_media['satisfacao_media_positivo'], color='orange', label='Positivo')
plt.bar(satisfacao_media['intervalo_tempo'], satisfacao_media['satisfacao_media_negativo'], color='purple', label='Negativo', bottom=satisfacao_media['satisfacao_media_positivo'])
plt.xlabel('Intervalo de Tempo de Resposta')
plt.ylabel('Satisfação Média')
plt.title('Satisfação Média por Intervalo de Tempo de Resposta')
plt.legend()
plt.show()
```

## Serie temporal

```python
dados_ultima_data['data_atualizacao'] = pd.to_datetime(dados_ultima_data['data_atualizacao'])
dados_ultima_data['ano'] = dados_ultima_data['data_atualizacao'].dt.year
dados_ultima_data['mes'] = dados_ultima_data['data_atualizacao'].dt.month
dados_agrupados = dados_ultima_data.groupby(['ano', 'mes', 'score'])['tempo_de_resposta'].median().unstack()


for ano, dados_ano in dados_agrupados.groupby(level=0):
    plt.figure(figsize=(10, 6))
    plt.plot(dados_ano.index.get_level_values('mes'), dados_ano['positivo'], label='Positivo', marker='o', color='orange')
    plt.plot(dados_ano.index.get_level_values('mes'), dados_ano['negativo'], label='Negativo', marker='s', color='purple')
    plt.xlabel('Mês')
    plt.ylabel('Mediana do Tempo de Atendimento')
    plt.title(f'Mediana do Tempo de Atendimento por Mês - {ano}')
    plt.xticks(range(1, 13))
    plt.legend()
    plt.grid(True)
    plt.show()

```

##Conclusão

Satisfação Média em relação ao Intervalo de Tempo:

A satisfação média dos usuários é significativamente maior para os tickets que foram resolvidos dentro do prazo de até 1 dia (94%) e até 3 dias (92%, quando segmentamos mínima é até 3 dias) em comparação com aqueles que levaram mais de 3 dias para serem resolvidos (24%, quando segmentação mínima é até 3 dia). Isso sugere que a rapidez na resolução dos tickets tem um impacto positivo na satisfação. Portanto, é importante que continuemos priorizando e agilizando o processo de resolução de tickets para melhorar a experiência e sempre termos a satisfação positiva

Proporção de Tickets Avaliados como "Positivos" e "Negativos":

A proporção de tickets avaliados como "positivos" é muito maior (92%) em comparação com a proporção de tickets avaliados como "negativos" (78%). Isso indica que a maioria dos usuários tende a avaliar positivamente o atendimento recebido, o que é uma boa notícia!
